## Correlations

```python
# the heat map shows that some of the predictors are important to determine the price (garage, fireplace, year built)
# also, some of the variables are strongly correlated to each other, which may cause collinearity issues
def correlation_heat_map(df):
    corrs = df.corr()
    fig, ax = plt.subplots(figsize=(18, 18))
    mask = np.zeros_like(corrs, dtype=bool)
    mask[np.triu_indices_from(mask)] = True
    ax = sns.heatmap(corrs, mask=mask, annot=True, square = True)
    ax.set_xticklabels(ax.xaxis.get_ticklabels(), fontsize=12, rotation=90)
    ax.set_yticklabels(ax.yaxis.get_ticklabels(), fontsize=12, rotation=0)
    plt.show()
    return fig, ax

fig, ax = correlation_heat_map(df)
```

```python
#finding the top 20 absolute pearson correlation values
sorted_corr = df.corr().abs().unstack().sort_values(ascending = False,kind="quicksort")
sorted_corr[sorted_corr.values !=1][:20]
```

```python
#checking spearman ranking lo quickly look at non-linear correlations
def spearman_heat_map(df):
    corrs = df.corr(method = 'spearman')
    fig, ax = plt.subplots(figsize=(18, 18))
    mask = np.zeros_like(corrs, dtype=bool)
    mask[np.triu_indices_from(mask)] = True
    ax = sns.heatmap(corrs, mask=mask, annot=True, square = True)
    ax.set_xticklabels(ax.xaxis.get_ticklabels(), fontsize=12, rotation=90)
    ax.set_yticklabels(ax.yaxis.get_ticklabels(), fontsize=12, rotation=0)
    plt.show()
    return fig, ax

fig, ax = spearman_heat_map(df)
```

# Modelling

### Train-test split

```python
df_dum = pd.get_dummies(df,drop_first = True)
```

```python
X = df_dum.drop(['SalePrice', 'YrSold'], axis = 1)
X_train = df_dum[df_dum.YrSold != 2010].drop(['SalePrice', 'YrSold'], axis = 1)
X_test = df_dum[df_dum.YrSold == 2010].drop(['SalePrice', 'YrSold'], axis = 1)
y_train = df[df.YrSold != 2010].SalePrice
y_test = df[df.YrSold == 2010].SalePrice
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

### Preprocessing and linear regression

```python
scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X.columns)
```

```python
# create a linear regression model instance
model = LinearRegression()

# get cross validated scores
scores = cross_val_score(model, X_train, y_train, cv=5)
print("Cross-validated training scores:", scores)
print("Mean cross-validated training score:", scores.mean())
# fit and evaluate the data on the whole training set
model.fit(X_train, y_train)
print("Training Score:", model.score(X_train, y_train))
# evaluate the data on the test set
print("Test Score:", model.score(X_test, y_test))
```

## Applying regularization

### Ridge

```python
# create a RidgeCV model instance
ridge = RidgeCV(alphas=np.logspace(-4, 4, 10), cv=5)
# fit the model
ridge.fit(X_train, y_train)
# get the best alpha
print('Best alpha:', ridge.alpha_)
# evaluate on the training set
print('Training score:', ridge.score(X_train, y_train))
# evaluate on the test set
print("Test Score:", ridge.score(X_test, y_test))
# cv
print("Mean cross-validated training score:",cross_val_score(ridge, X_train, y_train, cv=5).mean())
```

### Lasso

```python
# create a LassoCV model instance
lasso = LassoCV(alphas=np.logspace(-4, 4, 10), cv=5, n_jobs=-1)
# fit the model
lasso.fit(X_train, y_train)
# get the best alpha
print('Best alpha:', lasso.alpha_)
# evaluate on the training set
print('Training score:', lasso.score(X_train, y_train))
# evaluate on the test set
print("Test Score:", lasso.score(X_test, y_test))
# cv
print("Mean cross-validated training score:",cross_val_score(lasso, X_train, y_train, cv=5).mean())
best_lasso = Lasso(alpha = lasso.alpha_)
```

### ElasticNet

```python
# create an ElastinNetCV model instance
elasticnet = ElasticNetCV(alphas=np.logspace(-4, 4, 10), 
                     l1_ratio=np.array([0.00001, .1, .5, .7, .9, .95, .99, 1]),
                     cv=5, max_iter=100000)
# fit the model
elasticnet.fit(X_train, y_train)
# get the best alpha
print('Best alpha:', elasticnet.alpha_)
# get the best l1-ratio
print('Best l1-ratio:', elasticnet.l1_ratio_)
# evaluate on the training set
print('Training score:', elasticnet.score(X_train, y_train))
# evaluate on the test set
print("Test Score:", elasticnet.score(X_test, y_test))
print("Mean cross-validated training score:",cross_val_score(elasticnet, X_train, y_train, cv=5).mean())
```

## Coefficients importance

As shown above, Lasso regularization performs best. Plotting top features by importance below

```python
# collect the model coefficients in a dataframe
df_coef = pd.DataFrame(lasso.coef_, index=X_train.columns,
                       columns=['coefficients'])
# calculate the absolute values of the coefficients
df_coef['coef_abs'] = df_coef.coefficients.abs()
df_coef.head()
```

```python
#sorting by absolute coefficient and creating a df with only the top 20 for ease of visualization
df_coef_top20 = df_coef.sort_values(by = 'coef_abs', ascending = False).head(30)
df_coef_top20['coefficients'].sort_values().plot(kind='barh',figsize = (10,10));
```

## Non-linear models

Attempting to find models with higher variance and use ensemble methods to optimize test scores

### KNN regressor with gridsearch

```python
params = {'n_neighbors': list(range(2,15)),
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan']}

knn = KNeighborsRegressor(n_jobs = -2)

# create your gridsearch object
knn_gs = GridSearchCV(knn,
                             params,
                             n_jobs=-2, 
                             cv=5, 
                             verbose=1)
knn_gs.fit(X_train,y_train)
```

```python
#model is overfitting, although r2 score on test set remains above 0.80 
print(knn_gs.best_params_)
print(knn_gs.score(X_train,y_train))
print(knn_gs.score(X_test,y_test))
print(cross_val_score(knn_gs.best_estimator_, X_train, y_train, cv=5).mean())
best_knn = knn_gs.best_estimator_
```

### Decision tree

```python
params_dtree = {'max_depth' : [None] + list(range(1,21)),
               'max_features': [None, 'auto', 'sqrt'],
               'criterion': ['mse','mae','poisson'],
               'min_samples_leaf': list(range(1,6))}
dtree = DecisionTreeRegressor()
dtree_gs = GridSearchCV(dtree,params_dtree,
                              n_jobs=-2, 
                              cv=5, 
                              verbose=1, )
dtree_gs.fit(X_train,y_train)
```

```python
print(dtree_gs.best_params_)
print(dtree_gs.score(X_train,y_train))
print(dtree_gs.score(X_test,y_test))
print(cross_val_score(dtree_gs.best_estimator_, X_train, y_train, cv=5).mean())
best_dtree = dtree_gs.best_estimator_
```

## Ensembles

### Bagged tree

Bagging retains high variance but slightly improves 

```python
params = {'max_features': [0.5, 0.7, 1.0],
          'max_samples': [0.5, 0.7, 1.0],
          'n_estimators': [2, 5, 10, 20, 50]}

bagging = BaggingRegressor(n_jobs = -2)

bagging_gs = GridSearchCV(bagging,params,
                              n_jobs=-2, 
                              cv=5, 
                              verbose=1)
bagging_gs.fit(X_train,y_train)
```

```python
bagging_gs_best = bagging_gs.best_estimator_
print(bagging_gs.best_params_)
print(bagging_gs_best.score(X_train, y_train))
print(bagging_gs_best.score(X_test, y_test))
print(cross_val_score(bagging_gs_best, X_train, y_train, cv = 5).mean())
```

### Random forest

Random forest performs slightly worse than bagged trees on the 2010 data

```python
params = {'max_depth' : [None] + [3,12],
               'criterion': ['mse','mae'],
               'min_samples_leaf': [1,4],
            'n_estimators': [50, 100, 200],
            'max_features': ['auto','sqrt']}

rf = RandomForestRegressor(random_state=1)
rf_gs= GridSearchCV(rf , params, cv = 5, verbose = 1, n_jobs = -2)
rf_gs.fit(X_train,y_train)
```

```python
rf_gs_best = rf_gs.best_estimator_
print(rf_gs.best_params_)
print(rf_gs_best.score(X_train, y_train))
print(rf_gs_best.score(X_test, y_test))
print(cross_val_score(rf_gs_best, X_train, y_train, cv = 5).mean())
```

### Gradient boosting

Gradient boosting has the best test score so far

```python
params = {'n_estimators':[50,100,200],
          "min_samples_leaf": [0.05,0.01],
                "max_depth":[1,2,3,4],
                "max_features":["log2",'sqrt'],
                'learning_rate': [0.05,0.1, 0.01]
              }
gboost = GradientBoostingRegressor(random_state=1)
gboost_gs = GridSearchCV(gboost, params, cv = 5, verbose = 1, n_jobs = -2)
gboost_gs.fit(X_train,y_train)
```

```python
gboost_gs_best = gboost_gs.best_estimator_
print(gboost_gs.best_params_)
print(gboost_gs_best.score(X_train, y_train))
print(gboost_gs.score(X_test, y_test))
print(cross_val_score(gboost_gs, X_train, y_train, cv = 5).mean())
```

## Voting

Will group all the models in an ensemble through voting

```python
voting = VotingRegressor([('dt_bagging', bagging_gs_best ), ('knn', best_knn), ('rf',rf_gs_best),('gradboost',gboost_gs_best), ('bagging',bagging_gs_best),('lasso', best_lasso)], n_jobs = -2)
voting.fit(X_train,y_train)
```

```python
print(voting.score(X_train, y_train))
print(voting.score(X_test, y_test))
```

## 2. Determine any value of *changeable* property characteristics unexplained by the *fixed* ones.

---

Now that you have a model that estimates the price of a house based on its static characteristics, we can move forward with part 2 and 3 of the plan: what are the costs/benefits of quality, condition, and renovations?

There are two specific requirements for these estimates:
1. The estimates of effects must be in terms of dollars added or subtracted from the house value. 
2. The effects must be on the variance in price remaining from the first model.

The residuals from the first model (training and testing) represent the variance in price unexplained by the fixed characteristics. Of that variance in price remaining, how much of it can be explained by the easy-to-change aspects of the property?

---

**Your goals:**
1. Evaluate the effect in dollars of the renovatable features. 
- How would your company use this second model and its coefficients to determine whether they should buy a property or not? Explain how the company can use the two models you have built to determine if they can make money. 
- Investigate how much of the variance in price remaining is explained by these features.
- Do you trust your model? Should it be used to evaluate which properties to buy and fix up?

```python
renov_cols =[col for col in house.columns if col not in fixed_cols]
renov_df=house[renov_cols + ['SalePrice','YrSold']]
renov_df.drop('Id', inplace = True,axis = 1)
```

```python
#creating a column for residuals
concat_resids = pd.concat([resids_train,resids_test])
renov_df['residuals'] = concat_resids
renov_df['residuals'].head()
```

```python
df_dum = pd.get_dummies(renov_df,drop_first = True)
X = df_dum.drop(['SalePrice', 'YrSold', 'residuals'], axis = 1)
X_train = df_dum[df_dum.YrSold != 2010].drop(['SalePrice', 'YrSold', 'residuals'], axis = 1)
X_test = df_dum[df_dum.YrSold == 2010].drop(['SalePrice', 'YrSold', 'residuals'], axis = 1)
y_train = df_dum[df_dum.YrSold != 2010].residuals
y_test = df_dum[df_dum.YrSold == 2010].residuals
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X.columns)
```

```python
lasso = LassoCV(alphas=np.logspace(-4, 4, 10), cv=5, n_jobs=-1)
lasso.fit(X_train, y_train)

print('Best alpha:', lasso.alpha_)
print('Training score:', lasso.score(X_train, y_train))
print("Test Score:", lasso.score(X_test, y_test))
print("Mean cross-validated training score:",cross_val_score(lasso, X_train, y_train, cv=5).mean())
```

```python
# collect the model coefficients in a dataframe
df_coef = pd.DataFrame(lasso.coef_, index=X_train.columns,
                       columns=['coefficients'])
# calculate the absolute values of the coefficients
df_coef['coef_abs'] = df_coef.coefficients.abs()
df_coef.head()
```

```python
#sorting by absolute coefficient and creating a df with only the top 20 for ease of visualization
df_coef_top20 = df_coef.sort_values(by = 'coef_abs', ascending = False).head(30)
df_coef_top20['coefficients'].sort_values().plot(kind='barh',figsize = (10,10));
```

```python
import scikitplot as skplt
from matplotlib.colors import ListedColormap
cmap = ListedColormap(sns.color_palette("husl", len(gs.classes_)))
from sklearn.metrics import classification_report, confusion_matrix
```

```python
def fitter(model, X_train, y_train, X_test, y_test, cv_score=True):
    """
    Takes a model, training and test sets as inputs and evaluated the model on both 
    reporting scores, confusion matrix and classification report.
    """

    print(model.score(X_train, y_train))
    if cv_score:
        print(cross_val_score(model, X_train, y_train, cv=5).mean())
    print(f'test score: {model.score(X_test, y_test)}')
    print()
    skplt.metrics.plot_confusion_matrix(y_train, model.predict(X_train), labels=[1, 0], figsize=(6,6))
    plt.show()
    print()
    print(classification_report(y_train, model.predict(X_train)))
    print()
    skplt.metrics.plot_confusion_matrix(y_test, model.predict(X_test), labels=[1, 0], figsize=(6,6))
    plt.show()
    print()
    print(classification_report(y_test, model.predict(X_test)))
```

```python
#fitting a simple logistic regression, test set contains only 3 abnormal sales, of which 2 were predicted wrongly
model = LogisticRegression(max_iter = 1000)
model.fit(X_train, y_train)
fitter(model,X_train,y_train,X_test,y_test)
```
