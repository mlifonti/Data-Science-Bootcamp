## 7. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the base statistics of all their pokemon.

Your function will:

1. Accept the `players` dictionary, `pokedex` dictionary, and a player_id as arguments.
2. For the specified player_id, look up that player's pokemon and their level(s).
3. Find and aggregate the attack and defense values for each of the player's pokemon from the `pokedex` dictionary.
4. Print "[player name]'s power is [player power].", where the player power is the sum of the base statistics for all of their pokemon.
5. Return the player's power value.

Print out the pokemon power for each of your players.

````python
def calc_player_power(players, pokedex, player_id):
    '''Takes the players and pokedex dictionaries, and a player id to retrieve the sum of attack and defense stats of all of his pokemon'''
    power_list = []
    for id, stats in players[player_id]['player_pokemon'].items():
        power_list.append(stats['attack'] + stats['defense'] + stats['special_attack'] + stats['special_defense'])
    return sum(power_list)
print('Player 1 power is: {}'.format(calc_player_power(players, pokedex, 1)))
print('Player 2 power is: {}'.format(calc_player_power(players, pokedex, 2)))
````

## 8. Load a pokedex file containing all the pokemon

---

### 8.1

While you were putting together the prototype code, your colleagues were preparing a dataset of Pokemon and their attributes. (This was a rush job, so they may have picked some crazy values for some...)

The code below loads information from a comma separated value (csv) file. You need to parse this string into a more useable format. The format of the string is:

- Rows are separated by newline characters: \n
- Columns are separated by commas: ,
- All cells in the csv are double quoted. Ex: "PokedexNumber" is the first cell of the first row.


Using for-loops, create a list of lists where each list within the overall list is a row of the csv/matrix, and each element in that list is a cell in that row. Additional criteria:

1. Quotes are removed from each cell item.
2. Numeric column values are converted to floats.
3. There are some cells that are empty and have no information. For these cells put a -1 value in place.

Your end result is effectively a matrix. Each list in the outer list is a row, and the *j*th elements of the list together form the *j*th column, which represents a data attribute. The first three lists in your pokedex list should look like this:

    ['PokedexNumber', 'Name', 'Type', 'Total', 'HP', 'Attack', 'Defense', 'SpecialAttack', 'SpecialDefense', 'Speed']
    [1.0, 'Bulbasaur', 'GrassPoison', 318.0, 45.0, 49.0, 49.0, 65.0, 65.0, 45.0]
    [2.0, 'Ivysaur', 'GrassPoison', 405.0, 60.0, 62.0, 63.0, 80.0, 80.0, 60.0]

````python
# Code to read in pokedex info
raw_pd = ''
pokedex_file = 'pokedex_basic.csv'
with open(pokedex_file, 'r') as f:
    raw_pd = f.read()
    
# the pokedex string is assigned to the raw_pd variable
````

````python
#removing quote marks from data point
raw_pd = raw_pd.replace('"','')

pd_cleaned = []

#splitting the string into a list of lists
for row in raw_pd.split('\n'):
    dummy= []
    #converting numeric values into float, adding to a dummy list in each iteration, which is then appended to the main list
    for cell in row.split(','):
        #each value is converted to float if it is a number
        if cell.isnumeric() == True:
            dummy.append(float(cell))
        else:
            dummy.append(cell)
    pd_cleaned.append(dummy)

      
print(pd_cleaned[0:3])
````

## 9. Write a function to generate the full pokedex

---

Write a function that recreates the pokedex you made before, but with the data read in from the full pokemon file. Create a unique key value for each entry in the pokemon dictionary.

Your function should:

1. Take the parsed pokedex information you created above as an argument.
2. Return a dictionary in the same format as your original pokedex you created before containing the information from the parsed full pokedex file.

To test the function, print out the pokemon with id = 100.

````python
def create_pokedex(pd_cleaned):
    '''Generate full pokedex from parsing the cleaned csv'''
    #setting first list as header and the rest as data
    header = pd_cleaned[0]
    data = pd_cleaned[1:]
    pokedex_full = {}
    
    # Creating the dictionary by iterating
    for row in data:
        pokemon_stat = {}
        #using enumerate to have an index i to select the appropriate head and row.
        for i, pok in enumerate(row[1:],start =1):
            pokemon_stat[header[i]] = row[i]
        #row 0 is the pokemon id
        pokedex_full[row[0]] = pokemon_stat
    return pokedex_full    

pokedex_full = create_pokedex(pd_cleaned)
````

## 10. Write a function to generate a "filtered" pokedex
---
Your function should:
1. Take the parsed pokedex information you created above as an argument.
1. Take a dictionary as a parameter with keys matching the features of the Pokedex, filtering by exact match for string type values, and/or filter continuous variables specified by a value that is greater than or equal to the value of the corresponding dictionary key parameter.
1. Return multiple elements from the Pokedex

Example:

```python

# Only filter based on parameters passed
filter_options = {
    'Attack':   25,
    'Defense':  30,
    'Type':     'Electric'
}

# Return records with attack >= 25, defense >= 30, and type == "Electric"
# Also anticipate that other paramters can also be passed such as "SpecialAttack", "Speed", etc.
filtered_pokedex(pokedex_data, filter=filter_options)

# Example output:
# [{'Attack': 30.0,
#  'Defense': 50.0,
#  'HP': 40.0,
#  'Name': 'Voltorb',
#  'SpecialAttack': 55.0,

#  'SpecialDefense': 55.0,
#  'Speed': 100.0,
#  'Total': 330.0,
#  'Type': 'Electric'},
#  ... etc
#  ]

```



````python
def filter_pokedex(pokedex_full, filter_options):
    filtered = []
    #checking if filter_options is valid, printing the invalid key
    for key in filter_options:
        if key not in list(pokedex_full.values())[0].keys():
            print(f'{key} is an invalid input, choose appropriate filter options')
            return None

    
    for key,value in filter_options.items():
        #if name is in the filter options, its value should be string and there should not be other options
        if key == 'Name' and len(filter_options.keys()) == 1:
            if type(filter_options[key]) == str:
                filtered = [{pok_id: pok_stats for pok_id, pok_stats in pokedex_full.items() if pok_stats['Name'] == filter_options['Name']}]
            else:
                print(f'{filter_options[key]} should be string')
                return None
        elif key == 'Name' and len(filter_options.keys()) > 1:
            print('Only specify name key if searching by name')
            return None
        
        #check if type is in the filter options
        elif 'Type' in filter_options.keys():
            if type(filter_options[key]) == str:
                filtered=[{pok_id: pok_stats for pok_id, pok_stats in pokedex_full.items() if pok_stats['Type'] == filter_options['Type']}]

                #continues if there are other filter options 
                for k in filter_options:
                    if type(filter_options[k]) == int or type(filter_options[k]) == float:

                        #iterates through the existing filtered list to check if there are pokemons with stats >= the numeric options
                        filtered = [{pok_id: pok_stats for pok_id, pok_stats in _dict.items() if pok_stats[k] >= filter_options[k]} for _dict in filtered]
        else:
            filtered=[{pok_id: pok_stats for pok_id, pok_stats in pokedex_full.items()}]
            for k in filter_options:
                    if type(filter_options[k]) == int or type(filter_options[k]) == float:

                        filtered = [{pok_id: pok_stats for pok_id, pok_stats in _dict.items() if pok_stats[k] >= filter_options[k]} for _dict in filtered]
    if filtered == []:
        return 'None found'
    else:
        return filtered


filter_options = {
    'Attack':   25,
    'Defense':  30,
    'Type':     'Electric'
}
                   
print(filter_pokedex(pokedex_full,filter_options))
````

### 11.1

What is the population mean and standard deviation of the "Total" attribute for all characters in the Pokedex?



````python
#Mean using python:
#Creating a list containing all totals in the pokedex, summing them with the sum function 
#and dividing by the length of the pokedex to find the mean.
pok_totals = [pok_stats['Total'] for pok_stats in pokedex_full.values()]
pok_mean = sum(pok_totals) /len(pokedex_full.keys())
pok_mean
````

````python
#checking result with numpy
np.mean(pok_totals)
````

````python
#standard deviation using python
pok_std = (sum([(tot - pok_mean) **2 for tot in pok_totals])/(len(pokedex_full.keys())))**0.5
pok_std
````

````python
#standard deviation using numpy
np.std(pok_totals)
````

The design team wants you to make the powerful Pokemon rare, and the weaklings more common. How would you set the probability $p_i$ of finding Pokemon *i* each time a player visits a gym?

Write a function that takes in a Pokedex number and returns a value $p_i$ for that character.

Hint: there are many ways you could do this. What do _you_ think makes sense? Start with simplifying assumptions: for example, you could assume that the probabilities of encountering any two Pokemon on one visit to a gym are independent of each other.

````python
## I am dividing the 721 pokemon in 5 categories, each with a total probability that a pokemon of that category will spawn.
#The categories contain pokemon based on their total power and are chosen arbitrarily (due to time constraints):
# - common (power 180 to 450), 65% chance of spawning
# - uncommon (power 450 to 550), 25% of spawning
# - rare (power 550 to 620), 8% of spawning
# - super-rare (power 620 to 700), 1.9% of spawning
# - legendary (power 700 to 780), 0.1% of spawning
#Each pokemon within the each category has the same probabilities of spawning
#This type of classification is often used in videogames and trading card games.
    
    
def probability(pokemon_id):
    '''The probability function will give the chance of spawning for the pokemon ID selected, which is determined by: probability of rarity group / number of pokemon in the rarity group '''
    common = []
    uncommon = []
    rare = []
    super_rare = []
    legendary = []
    for k,val in pokedex_full.items():
        if pokedex_full[k]['Total'] <450:
            common.append(pokedex_full[k]['Total'])
        elif pokedex_full[k]['Total'] >= 450 and pokedex_full[k]['Total'] <550:
            uncommon.append(pokedex_full[k]['Total'])
        elif pokedex_full[k]['Total'] >= 550 and pokedex_full[k]['Total'] <620:
            rare.append(pokedex_full[k]['Total'])
        elif pokedex_full[k]['Total'] >= 620 and pokedex_full[k]['Total'] <700:
            super_rare.append(pokedex_full[k]['Total'])
        else:
            legendary.append(pokedex_full[k]['Total'])
            
    if pokedex_full[pokemon_id]['Total'] in common:
        return 0.65/len(common)
    elif pokedex_full[pokemon_id]['Total'] in uncommon:
        return 0.25/len(uncommon)
    elif pokedex_full[pokemon_id]['Total'] in rare:
        return 0.08/len(rare)
    elif pokedex_full[pokemon_id]['Total'] in super_rare:
        return 0.019/len(super_rare)
    elif pokedex_full[pokemon_id]['Total'] in legendary:
        return 0.001/len(legendary)

probability(3)
````
